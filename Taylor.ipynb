```python
import math
import matplotlib.pyplot as plt
import numpy as np 
```

Generacja funkcji wyznaczonych przy pomocy Szeregów 

```python
h_sampl = 1
h_end = 90

h = [math.radians(i) for i in np.arange(0, h_end+1, h_sampl)]
f_mac = [1 + (-1) * (h[i])**2/2 for i in range(len(h))]
f_tay = [-1 * (-h[i]) + (-h[i])**3/(3*2) for i in range(90,-1, -1)] 
f_cos = [math.cos(i) for i in h]
```

```python
degrees = [i for i in range(91)] 
```

Wynik powyższych działań 

```python
plt.style.use('fivethirtyeight')

plt.plot(degrees, f_cos)
plt.plot(degrees, f_tay, linestyle='dashed')
plt.plot(degrees, f_mac, linestyle='dashed')
plt.title("Porównanie szeregów Taylora i Maclaurina \n do wartości rzeczywistych funkcji cos(x)")
plt.xlabel("Kąt [°]")
plt.ylabel("Wartość")
plt.grid(True)
plt.legend(["Cosinus", "Taylor", "Maclaurin"])
```

Szereg Taylora w większym przedziale zgadza się z funkcją cos(x) niż jest to w przypadku Maclaurina
