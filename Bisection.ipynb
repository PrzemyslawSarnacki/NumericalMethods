```python
import numpy as np
import math
import seaborn as sns
import matplotlib.pyplot as plt
```

Implementujemy metodę Gaussa-Seidel'a w postaci funkcji

```python
def bisection(f,a,b,N):
    # Check if using this method even makes sense 
    if f(a)*f(b) >= 0:
        print("Error")
        return None

    # We are copying values to new variables to avoid side effects
    a_ = a
    b_ = b
    # Iterate N times  
    for _ in range(1,N+1):
        # get average of range
        avg_val = (a_ + b_)/2
        # assign to variable in order to improve performance
        f_avg_val = f(avg_val)
        # Check a value
        if f(a_)*f_avg_val < 0:
            # if smaller than 0 assign avg to b
            a_ = a_
            b_ = avg_val
        # Check b value
        elif f(b_)*f_avg_val < 0:
            # if smaller than 0 assign avg to a
            a_ = avg_val
            b_ = b_
        # Check if we got exact root 
        elif f_avg_val == 0:
            print("Exactly this value is correct!")
            return avg_val
        else:
            print("Error")
            return None
    # after every iteration return average of a and b
    return (a_ + b_)/2
```

```python
def sine(x):
    """
    docstring
    """
    return math.sin(3*x) + math.cos(10*x)
```

```python
eee = bisection(sine, 1,2,25)
eee
```

```python

original_sine = np.array([sine(i) for i in np.arange(0,5,0.01)])
sns.lineplot(np.arange(0,5,0.01), original_sine)
plt.scatter(eee, 0, marker='o', s=25)
```
